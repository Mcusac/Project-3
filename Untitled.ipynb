```python
import pandas as pd
import sqlite3
```

```python
batting_db = pd.read_csv('Resources/Batting.csv')
batting_db
```

```python
pitching_db = pd.read_csv('Resources/Pitching.csv')
pitching_db
```

```python
batting_db.columns
```

```python
pitching_db.columns
```

```python
pitching_clean = pitching_db[['playerID', 'yearID', 'SO', 'BB','ERA', 'HR']]
pitching_clean
```

```python
pitcher_career = pitching_clean.groupby('playerID')['yearID'].agg(['first','last'])
pitcher_career
```

```python
pitching_final = pitching_clean.merge(pitcher_career, on='playerID')
pitching_final
```

```python
batting_clean = batting_db[['playerID', 'yearID', 'AB', 'H','SO','HR', '2B', '3B']]
batting_clean
```

```python
batting_clean['AVG'] = batting_clean['H'] / batting_clean['AB']
batting_clean['2Bper'] = batting_clean['2B'] / batting_clean['H']
batting_clean['3Bper'] = batting_clean['3B'] / batting_clean['H']
batting_clean['HRper'] = batting_clean['HR'] / batting_clean['H']
batting_clean = batting_clean.fillna(0)
batting_clean['1Bper'] = 1 - (batting_clean['2Bper'] + batting_clean['3Bper'] + batting_clean['HRper'])
batting_clean
```

```python
batter_career = batting_clean.groupby('playerID')['yearID'].agg(['first','last'])
batter_career
```

```python
batting_cleaner = batting_clean.merge(batter_career, on='playerID')
batting_cleaner
```

```python
batting_final = batting_cleaner[['playerID', 'SO', 'HR', 'AVG', '1Bper', '2Bper', '3Bper', 'HRper', 'first', 'last']]
batting_final
```

```python
database = "Baseball.sqlite"
conn = sqlite3.connect(database)
pitching_final.to_sql(name='Pitching', if_exists='replace', index='False', con=conn)
batting_final.to_sql(name='Batting', if_exists='replace', index='False', con=conn)
conn.close()
```
