```python
import pandas as pd
import sqlite3
```

```python
batting_db = pd.read_csv('Resources/Batting.csv')
batting_db
```

```python
pitching_db = pd.read_csv('Resources/Pitching.csv')
pitching_db
```

```python
batting_db.columns
```

```python
pitching_db.columns
```

```python
pitching_clean = pitching_db[['playerID', 'yearID', 'SO', 'BB','ERA', 'HR']]
pitching_clean
```

```python
batting_clean = batting_db[['playerID', 'yearID', 'AB', 'H','SO','HR', '2B', '3B']]
batting_clean
```

```python
batting_clean['AVG'] = batting_clean['H'] / batting_clean['AB']
batting_clean['2Bper'] = batting_clean['2B'] / batting_clean['H']
batting_clean['3Bper'] = batting_clean['3B'] / batting_clean['H']
batting_clean['HRper'] = batting_clean['HR'] / batting_clean['H']
batting_clean = batting_clean.fillna(0)
batting_clean['1Bper'] = 1 - (batting_clean['2Bper'] + batting_clean['3Bper'] + batting_clean['HRper'])
batting_clean
```

```python
batting_final = batting_clean[['playerID', 'SO', 'HR', 'AVG', '1Bper', '2Bper', '3Bper', 'HRper']]
batting_final
```

```python
database = "Baseball.sqlite"
conn = sqlite3.connect(database)
pitching_clean.to_sql(name='Pitching', if_exists='replace', index='False', con=conn)
batting_final.to_sql(name='Batting', if_exists='replace', index='False', con=conn)
conn.close()
```
